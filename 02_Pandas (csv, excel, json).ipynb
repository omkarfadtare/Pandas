#### 1) Read CSV
- `Syntax: pd.read_csv(Filepath/Filename)`
- If you want to skip some rows then you can pass flag as `skiprows = 1/2/3...`
- You can also do the same thing using `header = 1` It will produce the same output; meaning is my header is at 1
- In a case, When you dont have header then you can pass `header = none` by defaul then it will by default give column names as 0,1,2,3.. But if you want specific column names then, You can pass `header=none, names = ['EMP ID','Salary','Designation','Address']`
- If you want to create a dataframe of few rows of csv file then you can pass `nrows =3` It will then create a dataframe of only 3 rows excluding header.
- If you want to create a dataframe of specified columns only,then use flag `usecols = [list of column names]`.
- Sometimes a csv file consist of null values or a false data denoted by some random characters then you can replace them with `NaN`, To do so you can pass a list of those random characters or if you want it to apply on only on specific column then you can pass a dictionary as below:
  - `na_values = ['Not available', '?', '-', '0']`
  - `na_values = {'EMP ID':['?', '_'], 'Address': [-1, 0, 'Not available']}

```python
df = pd.read_csv('Movies.csv')
df
```

```python
df = pd.read_csv('Movies.csv',skiprows=2)
df
```

```python
df = pd.read_csv('Movies.csv', header = None)
df
```

```python
df = pd.read_csv('Movies.csv', usecols= ['star_rating','content_rating', 'genre' ])
df
```

#### 2) Create Excel
- `Syntax: df.to_excel('new_file_name','sheet_name')`
- If we want our dataframe to be added to some specific row or column then we have to pass `startrow and startcol` flag.
- `Convertors` flag are used to add some values or replace some values or replace null values in dataframe. For using convertor we first have to create a UDF and then we can pass that function in a dictionary format `convertors = key ; function : value ; column_names`.
- Writing two datafrmaes on two different sheets (refer below example).

```python
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```

```python
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
```

#### 1) Read Json
- The navtive format of a dataframe is a dictionary.
- We can convert a dataframe or a column/series data structure into dictionary format using df.to_dict() function
- Data is stored in a dictionary format in a JSON file.

```python
df = pd.DataFrame({'Number' : np.arange(11,21)})
df
```

```python
df['Square'] = df['Number'] ** 2
df['Cube'] = df['Number'] ** 3
df
```

```python
# Using UDF

def my_fun():
    data = df.to_dict()
    return data

my_fun()
```

```python
df.to_dict(orient = 'list')
```

```python
df.to_dict(orient = 'records') # List of dictionaries
```

```python
df.to_dict(orient = 'index')
```
